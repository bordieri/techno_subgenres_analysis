```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup
```

```python
#r = rq.get('http://www.dados.gov.br/dataset?q=seguranca') # fazendo a requisição

r = rq.get('https://www.adam-heaton.co.uk/playlists/techno-playlists/')

print(r.status_code)  # verificando se a requisição foi bem sucedida

print(r.headers['Content-Type'])  # verificando o tipo de dado 
                                  # retornado pelo request

print(r.text)  # verificando o conteúdo retornado
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
df = pd.DataFrame()

sections = soup.find_all('div',{'class':'col-lg-4'})

print(len(sections))

for section in sections:
    subgenre = section.find('h3').text
    links = section.find_all('a',{'class':'playlist-list'})
    print(len(links))
    
    for link in links:
        microgenre = link.getText()
        repo = link.get('href')
        print(repo)
        r_aux = rq.get('https://www.adam-heaton.co.uk' + repo)
        #print(r_aux)
        soup_aux = BeautifulSoup(r_aux.text, 'html.parser')
        
        playlists = []
        playlists.append(soup_aux.find('a',{'data-playlistname':microgenre + ' Spotify Playlist','class':''}))
        playlists.append(soup_aux.find('a',{'data-playlistname':microgenre + ' Spotify Archive Playlist','class':''}))
        playlists = list(filter(None, playlists))
        
        print(len(playlists))
        for playlist in playlists:
            playlist_type = playlist.getText()
            playlist_id = playlist.get('href').replace('https://open.spotify.com/playlist/','')
            
            df = df.append({'subgenre':subgenre,
                            'microgenre':microgenre,
                            'playlist_type':playlist_type,
                            'playlist_id':playlist_id},
                            ignore_index=True)
            print('playlist_type = ',playlist_type)
            print('playlist_type = ',playlist_id)
        
        #print(repo)

#repo_len.sort()
#print(repos)
df
```
